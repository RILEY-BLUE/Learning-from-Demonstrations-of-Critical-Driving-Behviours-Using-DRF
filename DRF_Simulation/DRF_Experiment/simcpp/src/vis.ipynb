```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
import csv
import json
```

```python
## read data from .csv files
def row_count(filename):
    with open(filename) as in_file:
        return sum(1 for _ in in_file)

in_filename = 'costMap.csv'
csvreader = csv.reader(open(in_filename, newline='\n'), quoting=csv.QUOTE_NONNUMERIC)

last_line_number = row_count(in_filename)
line_count = 0

header = []
xGrid = []
yGrid = []
totalCost = []

while line_count < last_line_number:
    #header = next(csvreader)
    #line_count += 1
    yGrid = next(csvreader)
    line_count += 1
    #header = next(csvreader)
    #line_count += 1
    xGrid = next(csvreader)
    #xGrid = [float(i) for i in xGrid]
    line_count += 1
    #header = next(csvreader)
    #line_count += 1
    for i in range(len(xGrid)):
        #print(xGrid)
        totalCost.append(next(csvreader))
        line_count += 1
    #print(totalCost)
    ## Visualise
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

    # Make data.
    X = np.asarray(xGrid, dtype=np.float64)

    Y = np.asarray(yGrid, dtype=np.float64)
    X, Y = np.meshgrid(X, Y)

    Z = np.array(totalCost, dtype=np.float64)
    
    np.reshape(Z, (len(X), len(Y)))


    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False)

    # Customize the z axis.
    #ax.set_zlim(-1.01, 1.01)
    #ax.zaxis.set_major_locator(LinearLocator(10))
    # A StrMethodFormatter is used automatically
    #ax.zaxis.set_major_formatter('{x:.02f}')

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    plt.show()

    xGrid = []
    yGrid = []
    totalCost = []
```
